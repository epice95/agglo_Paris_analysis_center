### Analyse de l'évolution du prix de l'immobilier à Paris 
Cette partie recense les informations importantes permettant d'évaluer l'évolution (et les potentielles perspectives d'évolutions) de la valeur de l'immobilier à Paris.
Le sujet se concentre sur le centre de l'agglomération de Paris.
Ici, on considère Paris, ainsi que la petite couronne.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
from sklearn.linear_model import LinearRegression

# On fait une liste avec les dvf stocker dans le dossier ../Inputs/DVF
folder_path = "C:/Users/NEZHA/OneDrive/Documents/EcoUrbaine_Project/Code/Inputs/DVF"
files = [
    "ValeursFoncieres-2019-S2.txt",
    "ValeursFoncieres-2020.txt",
    "ValeursFoncieres-2021.txt",
    "ValeursFoncieres-2022.txt",
    "ValeursFoncieres-2023.txt",
    "ValeursFoncieres-2024-S1.txt"
]

# On conserve que les colonnes utiles
cols_to_keep = ["Date mutation", "Valeur fonciere", "Code postal", "Commune", "Surface reelle bati", "Type local"]

# Codes postaux pour Paris + Petite Couronne
codes_postaux = list(range(75000, 75021)) + list(range(92000, 92141)) + list(range(93000, 93211)) + list(range(94000, 94291))

# On associe les départements aux codes postaux
departements = {
    75: list(range(75000, 75021)),
    92: list(range(92000, 92141)),
    93: list(range(93000, 93211)),
    94: list(range(94000, 94291))
}


df_list = []
chunksize = 50000  # Je réduit l'impact mémoire pour mon pc
# boucle on parcourt fichier par fichier avec des morceaux pour pas surcharger mon pc
for file in files:
    file_path = os.path.join(folder_path, file)
    print(f"Traitement de {file}...")
    
    for chunk in pd.read_csv(file_path, delimiter='|', encoding='utf-8', usecols=cols_to_keep, chunksize=chunksize, low_memory=False):
        chunk = chunk[chunk["Code postal"].isin(codes_postaux)]  # Filtrage des départements concernés
        chunk["Date mutation"] = pd.to_datetime(chunk["Date mutation"], errors='coerce')
        chunk["Année"] = chunk["Date mutation"].dt.year
        chunk["Valeur fonciere"] = pd.to_numeric(chunk["Valeur fonciere"].astype(str).str.replace(',', '.'), errors='coerce')
        chunk["Prix_m2"] = chunk["Valeur fonciere"] / chunk["Surface reelle bati"] # calcul du prix par metre carré
        
        # Filtrage des valeurs aberrantes, on a pris entre 1000 et 20000
        chunk = chunk[(chunk["Prix_m2"] >= 1000) & (chunk["Prix_m2"] <= 20000)]
        
        # On ajouter une colonne département avec .apply
        chunk["Département"] = chunk["Code postal"].apply(lambda x: next((dep for dep, codes in departements.items() if x in codes), None))
        
        df_list.append(chunk[["Année", "Code postal", "Département", "Prix_m2", "Type local"]])


# Fusion des données
df_final = pd.concat(df_list, ignore_index=True)

# on fait le calcul du prix moyen par an et département avec .mean et .unstack pour avoir un tableau compréhensible
prix_moyen_par_annee = df_final.groupby(["Année", "Code postal"])["Prix_m2"].mean().unstack()
prix_moyen_par_departement = df_final.groupby(["Année", "Département"])["Prix_m2"].mean().unstack()

# Graphique par code postal, c'est brut mais c'est pour vérifié que tout les codes postaux sont valides
prix_moyen_par_annee.plot(figsize=(12, 6))
plt.xlabel("Année")
plt.ylabel("Prix moyen au m² (€)")
plt.title("Évolution du prix moyen au m² en Île-de-France (Paris + Petite Couronne) par Code Postal")
plt.legend(title="Code Postal")
plt.grid()
plt.show()

# Graphique par département
prix_moyen_par_departement.plot(figsize=(12, 6))
plt.xlabel("Année")
plt.ylabel("Prix moyen au m² (€)")
plt.title("Évolution du prix moyen au m² en Île-de-France (Paris + Petite Couronne) par Département")
plt.legend(title="Département")
plt.grid()
plt.show()

# Affichage des résultats
print(prix_moyen_par_annee)
print(prix_moyen_par_departement)
```
